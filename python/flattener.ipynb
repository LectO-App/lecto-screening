```python
import json
```

```python
file = open("test.json")
data = json.load(file)
data = data["allExercises"]
len(data)
```

```python
result = [];

for question in data:
    for excercise in question["exercises"]:
        obj = question.copy()
        del obj["exercises"]
        obj["exercise"] = excercise.copy()
        result.append(obj)
    
```

```python
result
```

```python
final = open("flattened.json", "w")
json.dump(result, final, ensure_ascii=False)
```
